```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
food = ctrl.Antecedent(np.arange(0,11,1),'food')
service = ctrl.Antecedent(np.arange(0,11,1),'service')
tip = ctrl.Consequent(np.arange(10,25,1),'tip') 
food['poor']= fuzz.trimf(food.universe,[0,0,5])
food['average']= fuzz.trimf(food.universe,[0,5,10])
food['awesome']= fuzz.trimf(food.universe,[5,10,10])
service['poor']= fuzz.trimf(service.universe,[0,0,5])
service['average']= fuzz.trimf(service.universe,[0,5,10])
service['awesome']= fuzz.trimf(service.universe,[5,10,10])
tip['less']= fuzz.trimf(tip.universe,[10,14,18])
tip['normal']= fuzz.trimf(tip.universe,[14,18,22])
tip['much']= fuzz.trimf(tip.universe,[18,22,25])
service.view()
food.view()
tip.view()
tip['normal'].view()
rule1=ctrl.Rule(food['poor'] & service['poor'], tip['less'])
rule2=ctrl.Rule(food['poor'] & service['average'],tip['less'])
rule3=ctrl.Rule(food['poor'] & service['awesome'],tip['normal'])
rule4=ctrl.Rule(food['average'] & service['poor'],tip['less'])
rule5=ctrl.Rule(food['average'] & service['average'],tip['normal'])
rule6=ctrl.Rule(food['average'] & service['awesome'],tip['much'])
rule7=ctrl.Rule(food['awesome'] & service['poor'],tip['normal'])
rule8=ctrl.Rule(food['awesome'] & service['average'],tip['much'])
rule9=ctrl.Rule(food['awesome'] & service['awesome'],tip['much'])

tipping_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
tipping=ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['food'] = 6.2
tipping.input['service'] = 3.5
tipping.compute()
print(tipping.output['tip'])
tip.view(sim=tipping)
```

```python
from networkx.algorithms.bipartite.basic import color
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
windspeed = ctrl.Antecedent(np.arange(0.7,3.1,0.5),'windspeed')
temperature = ctrl.Antecedent(np.arange(20.4,41.2,5),'temperature')
rainfall = ctrl.Consequent(np.arange(0.2,9.1,1),'rainfall') 
windspeed['low']= fuzz.trimf(windspeed.universe,[0.7,0.7,2])
windspeed['medium']= fuzz.trimf(windspeed.universe,[0.7,2,3.1])
windspeed['high']= fuzz.trimf(windspeed.universe,[2,3.1,3.1])
temperature['low']= fuzz.trimf(temperature.universe,[20.4,20.4,30])
temperature['medium']= fuzz.trimf(temperature.universe,[20.4,30,41.2])
temperature['high']= fuzz.trimf(temperature.universe,[30,41.2,41.2])
rainfall['low']= fuzz.trimf(rainfall.universe,[0.2,0.2,5])
rainfall['medium']= fuzz.trimf(rainfall.universe,[0.2,5,9.1])
rainfall['high']= fuzz.trimf(rainfall.universe,[5,9.1,9.1])

windspeed.view()
temperature.view()
rainfall.view()
rainfall['medium'].view()
rule1=ctrl.Rule(windspeed['low'] & temperature['low'], rainfall['low'])
rule2=ctrl.Rule(windspeed['low'] & temperature['medium'],rainfall['medium'])
rule3=ctrl.Rule(windspeed['low'] & temperature['high'],rainfall['high'])
rule4=ctrl.Rule(windspeed['medium'] & temperature['low'],rainfall['low'])
rule5=ctrl.Rule(windspeed['medium'] & temperature['medium'],rainfall['high'])
rule6=ctrl.Rule(windspeed['medium'] & temperature['high'],rainfall['high'])
rule7=ctrl.Rule(windspeed['high'] & temperature['low'],rainfall['low'])
rule8=ctrl.Rule(windspeed['high'] & temperature['medium'],rainfall['medium'])
rule9=ctrl.Rule(windspeed['high'] & temperature['high'],rainfall['high'])

rain_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
rain=ctrl.ControlSystemSimulation(rain_ctrl)
rain.input['windspeed'] = 2
rain.input['temperature'] = 40
rain.compute()
print(rain.output['rainfall'])
rainfall.view(sim=rain)
```
